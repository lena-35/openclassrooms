# 🧪 Test de mon Laboratoire Deep Learning
Ce test vérifie que :
1. Python fonctionne.
2. La bibliothèque de calcul (Numpy) est là.
3. L'affichage graphique (Matplotlib) est actif.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Création de données bidon (simule 100 jours d'études)
jours = np.arange(1, 101)
# Simule une progression : on commence à 0 et on monte vers 100% de réussite
progression = 100 * (1 - np.exp(-jours/20))

# 2. Création du graphique
plt.figure(figsize=(10, 5))
plt.plot(jours, progression, color='dodgerblue', linewidth=3)

# 3. Habillage (pour que ce soit pro)
plt.title("Ma progression dans mes 289 cours OpenClassrooms", fontsize=14)
plt.xlabel("Jours d'apprentissage", fontsize=12)
plt.ylabel("Compétences acquises (%)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# 4. Affichage
plt.show()
```
